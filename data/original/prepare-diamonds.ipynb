```python
import numpy as np
import pandas as pd
target_name = 'price'
```

```python
raw_df = pd.read_csv("./diamonds.csv")
```

```python
raw_df.head(5)
```

```python
raw_df.info()
```

```python
raw_df.describe()
```

```python
# 元データをシャッフル
raw_df_shaffuled = raw_df.sample(frac=1).reset_index(drop=True)
raw_df_shaffuled = raw_df_shaffuled.drop('Unnamed: 0', axis=1)

# 一意のIDを付与
raw_df_shaffuled = raw_df_shaffuled.reset_index()

raw_df_shaffuled = raw_df_shaffuled.rename(columns={"index":"id"})
raw_df_shaffuled.head(10)
```

```python
from sklearn.model_selection import train_test_split
train_df , test_df =  train_test_split(raw_df_shaffuled , test_size=0.3, shuffle=False)
```

```python
print(train_df.info())
train_df.head()
```

```python
print(test_df.info())
test_df.head()
```

```python
# testの結果を分離

test_df_y = test_df.loc[:,["id",target_name]]
test_df_X = test_df.drop(columns=target_name)
```

```python
test_df_y.head()
```

```python
test_df_X.head()
```

```python
# 結果を保存する
train_df.to_csv("../input/train.csv", index=False)
test_df_X.to_csv("../input/test.csv", index=False)
test_df_y.to_csv("test_solution.csv", index=False)
```

```python
# sample_submission の作成
sample_submission = test_df_y
sample_submission[target_name] = 5000
sample_submission.head()
```
